Code for computing the topic distribution for each movie in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/data_annotated.csv")
df.head()
```

```python
df_categorized_movies = df.groupby(['movie', 'code']).size()
# convert count to percentage for each movie
df_categorized_movies = df_categorized_movies / df_categorized_movies.groupby(level=0).transform('sum') * 100
df_categorized_movies
```

```python
df_categorized_movies = df_categorized_movies.sort_values(ascending=False).reset_index(name='percentage')
df_categorized_movies
```

```python
# plot df_categorized_movies with sns
plt.figure(figsize=(15, 5))
plots = sns.barplot(x='movie', y='percentage', data=df_categorized_movies, hue='code')
plots.legend(loc='upper right')
plt.title('Topic Distribution for each movie')
plt.show()
```
